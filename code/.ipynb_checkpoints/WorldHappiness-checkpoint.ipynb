```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

## Read in the data

>##### 1) Drop unnecessary columns<br><br>2) Add year column to each dataset<br><br>3) Change the column names to match<br><br>4) Combine the datasets


```python
df_15 = pd.read_csv('2015.csv')

print(df_15.shape)
df_15.head()
```

```python
df_15.drop(columns = ['Region', 'Standard Error'], axis = 1, inplace = True)
```

```python
df_15.drop('Dystopia Residual', axis = 1, inplace = True)
```

```python
df_15.head()
```

```python
df_15.rename(columns = 
             {'Happiness Rank': 'Rank', 'Happiness Score': 'Score', 'Economy (GDP per Capita)': 'GDP', 
              'Health (Life Expectancy)': 'LifeExpectancy', 'Trust (Government Corruption)': 'GovTrust'}, inplace = True)

df_15.head()
```

```python
df_15['Year'] = 2015
```

```python
df_16 = pd.read_csv('2016.csv')

print(df_16.shape)
df_16.head()
```

```python
df_16.drop(columns = ['Region', 'Lower Confidence Interval', 'Upper Confidence Interval'], axis = 1, inplace = True)

df_16.head()
```

```python
df_16.rename(columns = 
             {'Happiness Rank': 'Rank', 'Happiness Score': 'Score', 'Economy (GDP per Capita)': 'GDP', 
              'Health (Life Expectancy)': 'LifeExpectancy', 'Trust (Government Corruption)': 'GovTrust'}, inplace = True)

df_16.drop('Dystopia Residual', axis = 1, inplace = True)

df_16.head()
```

```python
df_16['Year'] = 2016
```

```python
df_17 = pd.read_csv('2017.csv')

print(df_17.shape)
df_17.head()
```

```python
df_17.drop(columns = ['Whisker.high', 'Whisker.low'], axis = 1, inplace = True)
```

```python
df_17.head()
```

```python
df_17.rename(columns = 
             {'Happiness.Rank': 'Rank', 'Happiness.Score': 'Score', 'Economy..GDP.per.Capita.': 'GDP', 
              'Health..Life.Expectancy.': 'LifeExpectancy', 'Trust..Government.Corruption.': 'GovTrust'}, inplace = True)

df_17.drop('Dystopia.Residual', axis = 1, inplace = True)

df_17.head()
```

```python
df_17['Year'] = 2017
```

```python
df_18 = pd.read_csv('2018.csv')

print(df_18.shape)
df_18.head()
```

```python
df_18.rename(columns = 
             {'Overall rank': 'Rank', 'Country or region': 'Country', 'GDP per capita': 'GDP', 'Social support': 'Family',
              'Healthy life expectancy': 'LifeExpectancy', 'Freedom to make life choices': 'Freedom', 'Perceptions of corruption': 'GovTrust'}, inplace = True)

df_18.head()
```

```python
df_18['Year'] = 2018
```

```python
df_19 = pd.read_csv('2019.csv')

print(df_19.shape)
df_19.head()
```

```python
df_19.rename(columns = 
             {'Overall rank': 'Rank', 'Country or region': 'Country', 'GDP per capita': 'GDP', 'Social support': 'Family',
              'Healthy life expectancy': 'LifeExpectancy', 'Freedom to make life choices': 'Freedom', 'Perceptions of corruption': 'GovTrust'}, inplace = True)

df_19.head()
```

```python
df_19['Year'] = 2019
```

df_19.head()

```python
dfs = [df_15, df_16, df_17, df_18, df_19]

df = pd.concat(dfs,ignore_index=True)
```

```python
df.head() 
```

```python
df.shape
```

```python
df.to_csv('whole.csv')
```

```python
den_usa = df.loc[(df['Country'] == 'Denmark') | (df['Country'] == 'United States')]
```

```python
den_usa
```

```python
columns = ['Score', 'GDP', 'Freedom', 'GovTrust']

for col in columns:
    sns.catplot(x ='Year', y = col, hue = 'Country', data = df[(df['Country'] == 'Denmark') | (df['Country'] == 'United States')], kind = 'bar')
    #plt.title('Denmark vs. United States', size = 12)
    plt.savefig(f'den_usa{col}.png')
    plt.show()

plt.savefig('../images/den_usa.png')
    
plt.figure(figsize = (10, 8))
sns.lineplot(x = 'Year', y = 'Score', hue = 'Country', data = df[(df['Rank'] < 6) | (df['Country'] == 'United States')], palette = 'brg')
plt.xticks([2015, 2016, 2017, 2018, 2019], ['2015', '2016', '2017', '2018', '2019'])
plt.title('Top 5 Countries vs. United States from 2015 to 2019', size = 13)

plt.savefig('../images/top5_usa.png');

plt.figure(figsize = (10, 8))
sns.boxplot(x = 'Year', y = 'Score', data = df)
plt.title('Happiness Score Boxplots by Year', size = 14)
plt.savefig('../images/score.png');

plt.figure(figsize = (10, 8))
sns.boxplot(x = 'Year', y = 'Score', data = df[df['Rank'] < 40], palette = 'CMRmap')
sns.boxplot(x= 'Year', y = 'Score', data = df[df['Rank'] > 117], palette = 'Paired_r')
plt.title('Top 25% and Bottom 25% Happiness Score Boxplots', size = 14)
plt.savefig('../images/top_bottom.png');
```

```python
plt.figure(figsize = (10, 8))
sns.lineplot(x = 'Year', y = 'GDP', data = df[df['Rank'] < 40], palette = 'CMRmap')
sns.lineplot(x = 'Year', y = 'GDP', data = df[df['Rank'] > 117], palette = 'Paired_r')
plt.xticks([2015, 2016, 2017, 2018, 2019], [2015, 2016, 2017, 2018, 2019])
plt.title('Top 25% and Bottom 25% GDP Boxplots')
plt.savefig('../images/gdp_box.png');
```

```python
# https://seaborn.pydata.org/generated/seaborn.FacetGrid.html
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html

scaler = MinMaxScaler()

X = df.drop(columns = ['Country', 'Rank', 'Year'], axis = 1)

df_copy = df.copy()

df_copy[['Score', 'GDP', 'Family', 'LifeExpectancy', 'Freedom', 'GovTrust', 'Generosity']] = scaler.fit_transform(X)
```

```python
g = sns.FacetGrid(df_copy[['Country','Year','GovTrust', 'Score']], col = 'Country', col_wrap = 6)
g = g.map(plt.plot, 'Year', 'GovTrust', marker = '.')
g = g.map(plt.plot, 'Year', 'Score', marker = '^', color = 'green')
plt.title('Happiness Score vs. GovTrust', size = 14)
plt.savefig('../images/facet.png');
```

```python
plt.figure(figsize = (24, 4))
sns.pairplot(data = df, x_vars = df.drop(columns = ['Country', 'Rank', 'Score', 'Year'], axis = 1).columns, y_vars = ['Score'], height = 4, markers = '+')
plt.savefig('../images/to_score.png');
```
